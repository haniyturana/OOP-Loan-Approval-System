```python
import importlib
import step2_Pt2_module
importlib.reload(step2_Pt2_module)
from step2_Pt2_module import CreditScoreCustomer

print("--- Welcome to Credit Card Defaulter Check System ---")

total_apps = 0
approved_apps = 0

while True:
    try:
        print("\n" + "="*30)
        print("    Loan Application Form    ")
        print("="*30)

        # 1. Collect inputs
        ic = input("IC Number: ") 
        age = int(input("Age: "))
        income = float(input("Monthly Salary (RM): "))
        debt = float(input("Monthly Commitment (RM): "))
        score = int(input("Credit Score (300-850): "))
        loan_requested = float(input("Total Loan Request (RM): "))

        # 2. Verification
        print("\n" + "-"*30)
        print("Please confirm details:")
        print(f"IC: {ic}")
        print(f"Salary/Debt: RM{income} / RM{debt}")
        print(f"Credit Score: {score}")
        print("-"*30)
        
        sah = input("Proceed with evaluation? (y/n): ").strip().lower()
        
        if sah != 'y':
            print("\n>> Action undone. Data cleared.")
            continue 

        # 3. Process the application
        # Note: We pass 'ic' into the 'name' parameter of the class
        test_user = CreditScoreCustomer(ic, age, income, debt, score)
        status, reason = test_user.evaluate_loan(loan_requested)
        
        # Update summary stats
        total_apps += 1
        if status == "Approved":
            approved_apps += 1
        
        print(f"\nRESULT FOR IC: {ic}")
        print(f"Status: **{status}**")
        print(f"Sebab: {reason}")

    except ValueError:
        print("\n[Error]: Please enter numbers only for Age, Salary, Commitment, and Score.")

    # 4. Loop control
    lagi = input("\nDo you want to add another user? (y/n): ").strip().lower()
    if lagi != 'y':
        print("\n" + "="*30)
        print("      FINAL SESSION SUMMARY      ")
        print("="*30)
        print(f"Total Applications: {total_apps}")
        print(f"Approved: {approved_apps}")
        print(f"Rejected: {total_apps - approved_apps}")
        print("\nSystem Closed. Goodbye!")
        break
```
